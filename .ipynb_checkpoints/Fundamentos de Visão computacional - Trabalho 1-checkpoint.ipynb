## Introdução:
   
   Este estudo tem objetivo implementar algumas técnicas de segmentação de imagens como $1^a$ trabalho de implementação da disciplina de Fundamentos de Visão Computacional (INF01030/CMP197-2021/2), ministrada pelo Doutor Claudio Rosito Jung.

   A tarefa consistia em fazer a segmentação e a contagem de núcleos de células em imagens de microscopia. As imagens foram disponibilizadas no padrão RGB. Para realização desta tarefa esse trabalho foi dividido em duas partes:

1. Inicialmente se definiu quais seriam as imagens de entrada $I_f$ para as etapas subsequentes do processamento. As imagens foram escolhidas a partir de uma comparação visual entre as imagens do canal R com as imagens da diferença dos canais R-B. Nessa etapa deveria se escolher qual das duas imagens gera maior contraste entre os núcleos da células e o fundo. No final, $I_f$ deveria possuir apenas um canal. 
2. Definido as imagens de entrada $I_f$, aplicou se o processo de binarização para isolar o núcleo das células do fundo, essa etapa foi dividida em três partes:
    - Gerar os histogramas para as imagens e tentar identificar uma estratégia para calcular um limiar que separe a imagem do fundo, operação de ($\textit{threshold}$).
    - Aplicar a limiarização de Otsu para calcular o valor do threshold automaticamente. 
    - Após se compararia as imagens geradas por esses dois processo e se escolheria as imagens que tiveram melhor contraste para aplicação de um algoritmo de componentes conexos. Com os componentes conexos estraídos se calcularia algumas estatísticas desses componentes, como média e desvio padrão.
3. A partir das imagens binarizadas na etapa anterior se aplicaria a transformada de watershed para a extração das componentes conexas. Esta etapa foi dividida em cinco partes:
    - Obter o módulo do vetor gradiente $\parallel\bigtriangledown I_f\parallel$.
    - A partir das imagens binárias gerar um conjunto de marcadores para a transformada de watershed. Para tal deveria-se aplicar as operações morfológicas de fechamento e erosão em sequência utilizando um kernel de tamanho 3x3.
    - Aplica-se a transformada de watershed utiizando os marcadores utilizados anteriormente na imagem $\parallel\bigtriangledown I_f\parallel$.
    - Após aplicada a transformada de watershed obtenha-se as estatíticas dos núcleos das células.
    - Fazer uma análise visual dos resultados obtidos, comparando com a saída binária do item anterior.

# 2 Binarização

## Plotando o histograma para cada imagem

Nesta etapa calculamos os histogramas para cada imagem de entrada $I_f$ , para tentar identificar alguma estratégia que possa nos ajudar a definir automaticamente um valor que separe o núcleo das células para o fundo. As Figuras (a) composite_01.png, (b) composite_02.png, (c) composite_03.png e (d) composite_04.png são os histogramas da imagem de diferença entre os canais R e B.

```python
fig, axs = plt.subplots(2,2, figsize = (18,13))

axs[0][0].hist(images_difference[0].ravel(), bins = 255)
axs[0][1].hist(images_difference[1].ravel(), bins = 255)
axs[1][0].hist(images_difference[2].ravel(), bins = 255)
axs[1][1].hist(images_difference[3].ravel(), bins = 255)

titles = [['(a)', '(b)'], ['(c)', '(d)']]

for axss, titless in zip(axs,titles):
    for ax, title in zip(axss, titless):
        
        # Set title
        
        ax.set_title(title, fontsize = 32)
        
        # Set y label in scientific notation format
        
        ax.set_xlim(0,40)
        ax.ticklabel_format(axis='y',style='sci',scilimits=(0,0))
        ax.tick_params(axis='both',labelsize=18)
```

Observando os histogramas não conseguimos observar claramente dois picos que dividem a imagem. O mais próximo que conseguiu-se observar foi dois picos, mas ainda muito próximos, nos histogramas (b) e (d).

## Obtendo o valor do Thresholding de forma automâtica

Como não conseguimos observar anteriormente uma divisão muito clara nos histogramas das imagens. Por isso, para que se obtenha o valor do threshold para cada imagem de entrada $I_f$ automaticamente decidiu-se por usar a média da intensidade do histograma de cada imagem. Entre os motivos para a escolha dessa técnica está na sua facilidade e rápida implementação. O algoritmo completo da implementação dessa etapa pode ser visto no script $\textit{threshold.py}$.

```python
from Contents.threshold import threshold_mean
```

```python
thresholds = np.array([threshold_mean(
    images_difference[i]) for i in range(n_images)])
```

Abaixo podemos ver o valor do threshold calculado para cada imagem utilizando o método proposto.

```python
print(np.round(thresholds,2))
```

Em seguida utilizamos o valor do threshold calculado para produzir as imagens binárias.

```python
images_bi = []
for i in range(n_images):
    images_bi.append(images_difference[i] > thresholds[i])
images_bi = np.array(images_bi)
```

## Plotando as imagens binárias

Em baixo podemos cada uma das imagens de entrada $I_f$, Figuras (a), (c), (e) e (g), comparadas com suas versões binárias obtidas pelo método da intensidade média, Figuras (b), (d), (f) e (h).  

```python
fig, axs = plt.subplots(4,2, figsize = (10,18))

titles = [['(a)', '(b)'], ['(c)', '(d)'],
          ['(e)', '(f)'], ['(g)', '(h)']]

for index, axss, titless in zip(range(n_images),axs,titles):
    for image, ax, title in zip([
        images_difference[index], images_bi[index]],
        axss, titless):
        
        # Plotting images
        
        ax.imshow(image, cmap = 'gray')
        
        # Set title
        
        ax.set_title(title, fontsize = 24)
        
        # Remove axis
        
        ax.xaxis.set_major_locator(NullLocator())
        ax.yaxis.set_major_locator(NullLocator())
```

Observando as imagens consta-se que embora o valor de threshold calculado tenha conseguido destacar bem os núcleos das células, ele ainda gerou muito ruído, além de gerar um padrão circular na imagem. Isso pode ser um problema quando quisermos extrair os componentes conexos da imagem futuramente.

## Obtendo a imagem binária com método de Otsu

Nesta etapa repetiu-se o processo feito anteriormente, porém utilizou-se o método de Otsu para calcular o valor do threshold para cada imagem de entrada $I_f$. Utilizou-se a versão do algoritmo de Otsu implementado na biblioteca Skimage.

```python
from skimage.filters import threshold_otsu
```

```python
thresholds_otsu = np.array([threshold_otsu(
    images_difference[i]) for i in range(n_images)])
```

Abaixo pode-se ver o valor calculado do threshold para cada imagem de entrada $I_f$ pelo método de Otsu.

```python
print(thresholds_otsu)
```

Assim podemos obter a imagem binária.

```python
images_bi_otsu = []
for i in range(n_images):
    images_bi_otsu.append(images_difference[i] > thresholds_otsu[i])
images_bi_otsu = np.array(images_bi_otsu)
```

## Plotando as imagens binária com método de Otsu

Abaixo pode-se ver nas Figuras (a), (c), (e) e (g) as imagens de entrada, enquanto nas Figuras (b), (d), (f) e (h) pode se ver a imagem binária obtida pelo método de Otsu.

```python
fig, axs = plt.subplots(4,2, figsize = (10,18))

titles = [['(a)', '(b)'], ['(c)', '(d)'],
          ['(e)', '(f)'], ['(g)', '(h)']]

for index, axss, titless in zip(range(n_images),axs,titles):
    for image, ax, title in zip([
        images_difference[index], images_bi_otsu[index]],
        axss, titless):
        
        # Plotting images
        
        ax.imshow(image, cmap = 'gray')
        
        # Set title
        
        ax.set_title(title, fontsize = 24)
        
        # Remove axis
        
        ax.xaxis.set_major_locator(NullLocator())
        ax.yaxis.set_major_locator(NullLocator())
```

Observando visualmente as figuras nota-se que conseguiu-se delimitar nas imagens binárias o núcleo das células com bastante fidelidade, sem gerar muito ruído.

## Comparando as duas técnicas

Nas Figuras (a), (c), (e) e (g) pode-se ver as imagens binárias obtidas com o método da intensidade média, enquanto nas Figuras (b), (d), (f) e (h) vê-se as imagens binárias obtidas com o método de Otsu.

```python
fig, axs = plt.subplots(4,2, figsize = (10,18))

titles = [['(a)', '(b)'], ['(c)', '(d)'],
          ['(e)', '(f)'], ['(g)', '(h)']]

for index, axss, titless in zip(range(n_images),axs,titles):
    for image, ax, title in zip([
        images_bi[index], images_bi_otsu[index]],
        axss, titless):
        
        # Plotting images
        
        ax.imshow(image, cmap = 'gray')
        
        # Set title
        
        ax.set_title(title, fontsize = 24)
        
        # Remove axis
        
        ax.xaxis.set_major_locator(NullLocator())
        ax.yaxis.set_major_locator(NullLocator())
```

Comparando visualmente as imagens pode-se concluir que as imagens binárias geradas com o método de Otsu conseguiram não só captar bem o núcleo das células, como não apresentaram o ruído no fundo nem o padrão circular que aparece nas imagens que se usou a intensidade média para delimitar o threshold. Por este motivo selecionaremos as imagens binárias obtidas com o método de Otsu para as etapas seguintes do trabalho.

## Extraindo as componentes conexas

Nesta etapa são extraídas as componentes conexas a partir das imagens binárias. Para tal tarefa é usada a função $\textit{findContours}$ da biblioteca OpenCv. Abaixo pode-se ver o código utilizado.

```python
import cv2
```

```python
images_contours = []

for i in range(n_images):
    contours, hierarchy = cv2.findContours(
        np.uint8(images_bi_otsu[i]), 
        cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
    images_contours.append(contours)
```

## Eliminando componentes muito pequenos

Apesar do método de Otsu ter gerado imagens binárias com muito menos ruído que o outro método implementado, ainda assim quando aplicamos o método para delimitar as componentes conexas o resultado final ainda contém muitas componentes com areas pequenas de 3, 5 e 10 pixels de área. Para remover esses artefatos decidiu-se por desconsiderar toda a componente com uma área muito pequena, nesse caso menos que 100 pixels de área. O que é feito pelas linha abaixo.

Para maior compreensão da função $\textit{contours_filters}$ veja o script $\textit{filters.py}$ na pasta Contents dentro deste diretório.

```python
from Contents.filters import contours_filters
```

```python
images_contours_areas, images_contours_filters = contours_filters(
    images_contours,n_images)
```

## Marcando os componentes conexos nas imagens

Para termos uma visualização melhor dos resultados decidimos pintar nas imagens originais os contornos de cada componente. O que é feito pelo código logo abaixo.

```python
images_draw = []

yellow = (255,255,0)

for i in range(4):
    images_draw.append(
        cv2.drawContours(
            images[i].copy(), 
            images_contours_filters[i],
            -1,yellow,2))
    
images_draw = np.array(images_draw)
```

Observadas as imagens observa-se que conseguiu-se identificar boa parte dos núcleos das células. Todavia, ainda há alguns poucos núcleos que não foram identificados, especialmente nas Figuras (a) e (d). Isso ocorreu provavelmente porque esses núcleos são um pouco menores ou por terem ficado um pouco transparente nas imagens, se voltarmos nas imagens binarizadas vemos que esses núcleos foram case apagados nas imagens binárias.

## Estatísticas

Aqui calculamos algumas estatísticas dos núcleos das imagens. São elas: número de núcleos, tamanho do maior núcleo, tamanho do menor núcleo, média do tamanho dos núcleos, desvio padrão do tamanho dos núcleos e a mediana do tamanho dos núcleos.

```python
data = dict({
    'n_labels':[int(len(images_contours_areas[i])) for i in range(n_images)], 
    'max': [np.round(np.max(
        images_contours_areas[i]),2) for i in range(n_images)],
    'min': [np.round(np.min(
        images_contours_areas[i]),2) for i in range(n_images)],
    'mean': [np.round(np.mean(
        images_contours_areas[i]),2) for i in range(n_images)],
    'std': [np.round(np.std(
        images_contours_areas[i]),2) for i in range(n_images)],
    'median': [np.round(np.median(
        images_contours_areas[i]),2) for i in range(n_images)]
})
db = pd.DataFrame.from_dict(data, orient = 'index')
db.columns = [('Composite_0' + str(i)) for i in range(1,n_images+1)]
```

Abaixo podemos os dados estatísticos numa tabela. Observa-se que o número de núcleos nas imagem pode variar de 7 até 16. A imagem $\textit{Composite_3}$ foi a que teve a maior média no tamanho dos núcleos, enquanto a imagem $\textit{Composite_2}$ foi a que teve a menor média. Também se destaca o auto desvio padrão do tamanho das imagens comparado com a média das mesmas imagens, chegando 321 pixels na imagem $\textit{Composite_3}$ sendo que na mesma image é média é de 598 pixels, isso pode indicar dois coisas: que o tamanho dos núcleos variam muito dentro de cada imagem, ou que o n amostral de cada imagem é muito pequeno.

# 3 Transformada de watershed

## Aplicando o filtro de sobel para calcular o módulo do vetor gradiente $\parallel\bigtriangledown I_f\parallel$

Para aplicarmos a transformada de watershed presisamos primeiramente obter o módulo do vetor gradiente $\parallel\bigtriangledown I_f\parallel$. Para tal se optou pelo filtro de Sobel, a razão desta escolha é devido a capacidade deste filtro de identificar com bastante precisão os contornos na imagem, além de ser extremamente rápido. A versão utilizada do filtro de Sobel é a implementada pela biblioteca Skimage. O código utilizado pode ser visto nas linhas abaixo.

```python
from skimage.filters import sobel
```

```python
images_sobel = np.array(
    [np.abs(sobel(images_difference[i])) for i in range(n_images)])
```

## Plotando as imagens após aplicar o filtro de sobel

Abaixo pode se ver o módulo do vetor gradiente $\parallel\bigtriangledown I_f\parallel$, obtidas a partir do filtro de Sobel, para as imagens dos arquivos (a) composite_01.png, (b) composite_02.png, (c) composite_03.png e (d)composite_04.png.

```python
fig, axs = plt.subplots(2,2, figsize = (18,13))

axs[0][0].imshow(images_sobel[0], cmap = 'gray')
axs[0][1].imshow(images_sobel[1], cmap = 'gray')
axs[1][0].imshow(images_sobel[2], cmap = 'gray')
axs[1][1].imshow(images_sobel[3], cmap = 'gray')

titles = [['(a)', '(b)'], ['(c)', '(d)']]

for axss, titless in zip(axs,titles):
    for ax, title in zip(axss, titless):
        
        # Set title
        
        ax.set_title(title, fontsize = 32)
        
        # Remove axis
        
        ax.xaxis.set_major_locator(NullLocator())
        ax.yaxis.set_major_locator(NullLocator())
```

Observando as imagens nota-se que os contornos das células ficaram muito mais destacadas após a aplicação do filtro de Sobel.

# Aplicando as operações morfológicas

Depois de obter o módulo do vetor gradiente $\parallel\bigtriangledown I_f\parallel$ deve-se obter os marcadores para aplicar a transformada de watershed. Para tal se realizou duas operações morfológicas nas imagens binárias obtidas com o método de Otsu, são elas a de fechamento e a de erosão.

## Aplicando a operação de fechamento

Abaixo pode-se ver o código para a aplicação da operação de fechamento nas imagens binárias com um kernel 3x3. A versão utilizada tanto da operação de fechamento como de erosão é a implementada pela biblioteca Skimage.

```python
from skimage.morphology import closing, erosion 
```

```python
kernel = np.ones((3,3))
```

```python
images_close = np.array(
    [closing(images_bi_otsu[i], kernel) for i in range(n_images)])
```

## Plotando as imagens obtidas com a operação de fechamento

Abaixo pode se ver o resultado da aplicação da operação de fechamento nas imagens binárias dos arquivos (a) composite_01.png, (b) composite_02.png, (c) composite_03.png e (d)composite_04.png.

```python
fig, axs = plt.subplots(2,2, figsize = (18,13))

axs[0][0].imshow(images_close[0], cmap = 'gray')
axs[0][1].imshow(images_close[1], cmap = 'gray')
axs[1][0].imshow(images_close[2], cmap = 'gray')
axs[1][1].imshow(images_close[3], cmap = 'gray')

titles = [['(a)', '(b)'], ['(c)', '(d)']]

for axss, titless in zip(axs,titles):
    for ax, title in zip(axss, titless):
        
        # Set title
        
        ax.set_title(title, fontsize = 32)
        
        # Remove axis
        
        ax.xaxis.set_major_locator(NullLocator())
        ax.yaxis.set_major_locator(NullLocator())
```

## Aplicando a operação de erosão

Abaixo vê-se a aplicação da operação de erosão no resultado da etapa anterior, utilizando um kernel de 3x3.

```python
masks = np.array([erosion(images_close[i],kernel) for i in range(n_images)])
```

## Plotando as imagens com a operação de erosão

Abaixo pode se ver o resultado da aplicação da operação de erosão nas imagens produzidas pela etapa anterior nas imagens originarias dos arquivos (a) composite_01.png, (b) composite_02.png, (c) composite_03.png e (d)composite_04.png.

```python
fig, axs = plt.subplots(2,2, figsize = (18,13))

axs[0][0].imshow(masks[0], cmap = 'gray')
axs[0][1].imshow(masks[1], cmap = 'gray')
axs[1][0].imshow(masks[2], cmap = 'gray')
axs[1][1].imshow(masks[3], cmap = 'gray')

titles = [['(a)', '(b)'], ['(c)', '(d)']]

for axss, titless in zip(axs,titles):
    for ax, title in zip(axss, titless):
        
        # Set title
        
        ax.set_title(title, fontsize = 32)
        
        # Remove axis
        
        ax.xaxis.set_major_locator(NullLocator())
        ax.yaxis.set_major_locator(NullLocator())
```

## Aplicando a transformada de watershed

Obtendo-se tanto o módulo do vetor gradiente $\parallel\bigtriangledown I_f\parallel$ como os marcadores pelas operações morfológicas, pode-se agora aplicar a transformada de watershed. Abaixo pode se ver linhas de códigos implementadas para esta tarefa. A versão utilizada da transformada de watershed é implementada pela biblioteca Skimage.

```python
from skimage.segmentation import watershed
from scipy import ndimage as ndi
```

```python
images_labels = []
for i in range(n_images):
    markers,_ = ndi.label(masks[i])
    labels = watershed(
        images_sobel[i],
        markers,
        mask=masks[i])
    images_labels.append(labels)
images_labels = np.array(images_labels)
```

## Contando a área de cada segmento

Depois de aplicada a transformada de watershed e delimitado os núcleos das imagens, pode-se contar a área dos núcleos para que se possa obter as mesmas estatístiscas do item anterior. Abaixo vê-se o código para contar a área de cada núcleo.

```python
images_labels_count = [
    np.unique(images_labels[i], 
              return_counts = True)[1][1:] for i in range(n_images)]
```

Novamento eliminamos segmentos muito pequenos pelos motivos explicados anteriormente.

```python
images_labels_count_filter = [
    images_labels_count[i][
        images_labels_count[i] > area_minimum] for i in range(n_images)]
```

## Estatísticas

Novamente calculamos algumas estatísticas dos núcleos das imagens.

```python
data = dict({
    'n_labels':[
        int(len(images_labels_count_filter[i])) for i in range(n_images)], 
    'max': [
        np.round(np.max(images_labels_count_filter[i]),2) for i in range(n_images)],
    'min': [
        np.round(np.min(images_labels_count_filter[i]),2) for i in range(n_images)],
    'mean': [
        np.round(np.mean(images_labels_count_filter[i]),2) for i in range(n_images)],
    'std': [
        np.round(np.std(images_labels_count_filter[i]),2) for i in range(n_images)],
    'median': [
        np.round(np.median(images_labels_count_filter[i]),2) for i in range(n_images)]
})
db_watershed = pd.DataFrame.from_dict(data, orient = 'index')
db_watershed.columns = [('Composite_0' + str(i)) for i in range(1,n_images+1)]
```

Abaixo podemos ver as estatísticas calculadas dos núcleos identificados nas images usando a transformada de watershed. Comparando com a tabela anterior viu-se que não se modificou o número de núcleos identificados. Tanto a média do tamanho dos núcleos como o desvio padrão de cada imagem diminui um pouco.

# Conclusão:

Neste trabalho se aplicou técnicas de segmentação de imagem para identificação de núcleos de células em imagens obtidas por microscopia.
Primeiramente se fez uma análise visual do canal R das imagens e da imagem da diferença entre os canais R e B, observou-se que na imagem de diferença dos canais R e B é onde há maior contraste entre os núcleos das células e o fundo e por isso foi selecionada como imagem de entrada $I_f$ para as etapas posteriores do processamento. 

Em sequência se fez tanto a análise dos histogramas das imagens como se aplicou dois métodos diferentes para delimitar um valor limiar para separar o núcleo das células do fundo, no testes realizados a técnica que entregou o resultado mais satisfatório foi o método de Otsu. Com o valor do threshold calculado para cada imagem se fez a binarização das imagens e utilizou-se funcionalidades da biblioteca OpenCV para extrair as componentes conexas das imagens. Também se extraiu as componentes conexas da imagem utilizando a transformada de watershed, onde para tal se calculou o módulo do vetor gradiente $\parallel\bigtriangledown I_f\parallel$ utilizando o filtro de Sobel e delimitou-se marcadores a partir das operações morfológicas de fechamento e erosão nas imagens binarizadas. 

Identificado o núcleo das células de cada imagem, pode-se obter estatísticas tanto de média como desvio padrão para ambos os métodos utilizados. Os resultados indicam que conseguiu-se identificar de 7 a 16 núcleos de células das imagens, sendo a imagem $\textit{Composition_3}$ a que teve a maior média da área dos núcleos, assim como viu-se que há um grande desvio padrão em todas as imagens.
